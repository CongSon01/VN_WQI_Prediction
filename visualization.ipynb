```python
import pandas as pd
data = pd.read_excel('E:\WorkSpace\Research\ChuTuan\dulieuCLN_SongNhue_2024.xlsx', sheet_name='ketquaCLN')
data
```

```python
missing_values = data.isnull().sum()
print("Số lượng giá trị thiếu trong từng cột:")
print(missing_values)
```

```python
print("\nThống kê mô tả cơ bản cho các cột số liệu:")
descriptive_stats = data.describe()
print(descriptive_stats)
```

```python
# Thống kê trung bình, độ lệch chuẩn và các giá trị quan trọng cho các chỉ số môi trường
environmental_indicators = data[['Nhiệt độ', 'pH', 'Ôxy hòa tan (DO)', 'COD', 'BOD5 (20°C)', 
                                 'Amoni (NH4+ tính theo N)', 'Nitrit (NO2- tính theo N)', 
                                 'Nitrat (NO3- tính theo N)', 'Phosphat (PO43- tính theo P)']]
env_stats = environmental_indicators.describe()
print("\nThống kê mô tả cho các chỉ số môi trường:")
print(env_stats)
```

```python
# Thống kê và phân loại chất lượng nước dựa trên chỉ số WQI
data['WQI_Category'] = pd.cut(data['WQI'], bins=[0, 25, 50, 75, 100], labels=['Kém', 'Trung bình', 'Tốt', 'Rất tốt'])
wqi_summary = data['WQI_Category'].value_counts()
print("\nPhân loại chất lượng nước dựa trên WQI:")
print(wqi_summary)
```
